# Introduction to SQL

SQL (Structured Query Language) is a standard programming language used to manage and manipulate relational databases.

It is used to:
- Create and modify database structures
- Insert, update, and delete data
- Retrieve data using queries
- Maintain data integrity using constraints

In cybersecurity systems, SQL is used to store:
- Incident logs
- Vulnerability records
- Attack classifications
- Risk scores

In this notebook, we demonstrate:
- CREATE TABLE
- PRIMARY KEY
- INSERT statements
- Basic SELECT query

```python
import sqlite3
import pandas as pd
# Create in-memory database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

print("Database created successfully!")
```

## CREATE TABLE Statement

We create a table to store cybersecurity incident data.

```python
cursor.execute("""
CREATE TABLE CyberIncidents (
    Incident_ID INTEGER PRIMARY KEY,
    Timestamp TEXT,
    Source_Country TEXT,
    Target_System TEXT,
    Attack_Type TEXT,
    Severity_Level TEXT,
    Risk_Score INTEGER
);
""")

print("Table created successfully!")
```

## INSERT Statement

Now we insert cybersecurity incident records into the table.

```python
cursor.executemany("""
INSERT INTO CyberIncidents
(Incident_ID, Timestamp, Source_Country, Target_System, Attack_Type, Severity_Level, Risk_Score)
VALUES (?, ?, ?, ?, ?, ?, ?)
""", [
    (101, '2024-06-01 10:30:00', 'USA', 'Web Server', 'DDoS', 'High', 3),
    (102, '2024-06-02 14:20:00', 'India', 'Database', 'Phishing', 'Medium', 2),
    (103, '2024-06-03 09:15:00', 'Germany', 'Cloud VM', 'Ransomware', 'Critical', 4)
])

conn.commit()

print("Data inserted successfully!")
```

## SELECT Statement

Retrieve all records from the table.

```python
df = pd.read_sql_query("SELECT * FROM CyberIncidents", conn)
df
```

## Filtering Critical Incidents

Using WHERE clause to filter high severity threats.

```python
critical_df = pd.read_sql_query("""
SELECT Incident_ID, Attack_Type, Severity_Level
FROM CyberIncidents
WHERE Severity_Level = 'Critical';
""", conn)

critical_df
```

```python
conn.close()
print("Database connection closed.")
```
